# Kết quả thực nghiệm dữ liệu với PCA

```python
from pca import PCA_Processor ,print_results, visualize_comparison, visualize_pca_scatter
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report

# Khởi tạo đối tượng PCA_Processor
pca_processor = PCA_Processor(
    data_path='../data/framingham.csv',  # Đường dẫn đến file dữ liệu
    target_column='TenYearCHD',  # Cột mục tiêu
    n_components=2,  # Số chiều sau khi giảm chiều
    test_size=0.2,  # Tỷ lệ dữ liệu dùng để thử nghiệm
    random_state=42  # Hạt giống cho phép lặp lại các kết quả
)

# Định nghĩa danh sách các bộ phân loại
classifiers = [
    RandomForestClassifier(),
    LogisticRegression(),
    SVC(),
    KNeighborsClassifier(),
    GaussianNB(),
]

# Gọi hàm apply_pca_before_split
# print("\n\n===== PCA trước khi chia dữ liệu =====")
results_pca_before = pca_processor.apply_pca_before_split(classifiers)
print_results(results_pca_before)


# Gọi hàm apply_pca_after_split
# print("\n\n===== PCA sau khi chia dữ liệu =====")
results_pca_after = pca_processor.apply_pca_after_split(classifiers)
print_results(results_pca_after)

# Trực quan hóa cả hai trường hợp
visualize_pca_scatter(pca_processor)
```

# Trực quan hóa với UMAP

```python
from umap_function import UMAP_Processor, print_results
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Khởi tạo đối tượng UMAP_Processor
umap_processor = UMAP_Processor(
    data_path='../data/framingham.csv',  # Đường dẫn đến file dữ liệu
    target_column='TenYearCHD',  # Cột mục tiêu
    n_components=2,  # Số chiều sau khi giảm chiều
    test_size=0.2,  # Tỷ lệ dữ liệu dùng để thử nghiệm
    random_state=42  # Hạt giống cho phép lặp lại các kết quả
)

# Định nghĩa danh sách các bộ phân loại
classifiers = [
    RandomForestClassifier(),
    LogisticRegression(),
    SVC(),
    KNeighborsClassifier(),
    GaussianNB(),
]

# Gọi hàm apply_umap_before_split
print("\n\n===== UMAP trước khi chia dữ liệu =====")
results_umap_before = umap_processor.apply_umap_before_split(classifiers)
print_results(results_umap_before)

# Gọi hàm apply_umap_after_split
print("\n\n===== UMAP sau khi chia dữ liệu =====")
results_umap_after = umap_processor.apply_umap_after_split(classifiers)
print_results(results_umap_after)
```

# Phân cụm với Kmeans

```python
from kmeans_clustering import KMeansClusterer

# Khởi tạo
kmeans_processor = KMeansClusterer(
    data_path='../data/framingham.csv', 
    target_column='TenYearCHD'
)

# Tìm số lượng cụm tối ưu
optimal_clusters = kmeans_processor.find_optimal_clusters(method='silhouette')

# Thực hiện phân cụm
cluster_labels, kmeans_model = kmeans_processor.perform_clustering(optimal_clusters)

# Phân tích các cụm
df_clustered, cluster_summary = kmeans_processor.analyze_clusters(cluster_labels)
```
