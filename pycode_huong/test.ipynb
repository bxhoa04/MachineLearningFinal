```python
from umap_function import UMAP_Processor, print_results
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

# Khởi tạo đối tượng UMAP_Processor
umap_processor = UMAP_Processor(
    data_path='../data/framingham (1).csv',  # Đường dẫn đến file dữ liệu
    target_column='TenYearCHD',   # Cột mục tiêu
    n_components=2,  # Số chiều sau khi giảm chiều
    test_size=0.2,  # Tỷ lệ dữ liệu dùng để  thử nghiệm
    random_state=42  # Hạt giống cho phép lặp lại các kết quả
)

# Định nghĩa danh sách các bộ phân loại
classifiers = [
    GaussianNB(),
    KNeighborsClassifier(),
]

print("\n\n===== Hình ảnh trực quan hóa dữ liệu 2D sau khi áp dụng UMAP =====")
# Gọi hàm apply_umap_before_split
print("\n\n--- UMAP trước khi chia dữ liệu ---")
results_umap_before = umap_processor.apply_umap_before_split(classifiers)
# Gọi hàm apply_umap_after_split
print("\n\n--- UMAP sau khi chia dữ liệu ---")
results_umap_after = umap_processor.apply_umap_after_split(classifiers)
```

```python
#Kết quả trước khi áp dụng UMAP
print("\n\n===== Kết quả trước khi áp dụng UMAP =====")
results_before_umap = umap_processor.results_classider_before_umap(classifiers)
print_results(results_before_umap)
```

```python
#Kết quả sau khi áp dụng umap với n_components tối ưu
results_after_umap = umap_processor.apply_optimal_n_components(classifiers)
print("\n\n===== Kết quả sau khi áp dụng UMAP =====")
print_results(results_after_umap)
```
